```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

chromeDriver = "/Users/satty/Downloads/chromedriver"
url = "http://www.family.com.tw/marketing/inquiry.aspx"
driver = webdriver.Chrome(chromeDriver)
driver.get(url)
driver.find_element_by_xpath("//a[contains(@onclick,'台北市')]").click()
time.sleep(10)
driver.find_element_by_xpath("//a[contains(@onclick,'中正區')]").click()
time.sleep(3)
# go to next page
html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')
for page_num in soup.find_all(id = 'page_bu_content'):
    for li in page_num.find_all('li'):
        click_count = int(li.text)
print(click_count)
i = 0
while i < click_count:
    page_html_source = driver.page_source
    page_soup = BeautifulSoup(page_html_source, 'html.parser')
    td = page_soup.find_all('td', class_='graybox')
    for t in td:
        print(t.text)
    driver.find_element_by_xpath("//a[contains(@onclick,'next')]").click()
    i += 1
driver.close()
```
